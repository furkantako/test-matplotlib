## Scatter Plot

Verilerimize bakarak metre kareye göre fiyat tahmini yapmamız mümkün olur mu?


```python
import numpy as np
import matplotlib.pyplot as plt
#plt.rcParams['figure.figsize'] = [30, 30]

fig, ax = plt.subplots()

ax.scatter(evler['LotArea'], evler['SalePrice'], alpha=0.5)


ax.set(xlabel='alan', ylabel='fiyat',
       title='fiyat/alan oranı')

plt.show()
```

Bu grafikleri pandas ile üretmek de mümkün.

```python
ax2 = evler.plot.scatter(x='LotArea',y='SalePrice')
plt.show()
```

```python

ax2 = evler.plot.scatter(x='LotArea', y='SalePrice', c='NeighborhoodID',colormap='viridis', s=100, alpha=0.4)
plt.show()
```

Yukarıdaki görselimizi daha iyi anlayabilmek için belki parametreleri ne nokta boyutunu büyütmek için birşeyler ekleyebiliriz.
s=500

Notkaların şeffaflığını değiştirirsek, altta kalan veriyi de göreblime ihtimalimiz var.
alpha=0.2

```python
evler2 = evler[evler.NeighborhoodID ==9]
ax2 = evler2.plot.scatter(x='LotArea', y='SalePrice', c='NeighborhoodID',colormap='viridis')
plt.show()
```
