Kategorik verilerde görselleme fonksiyonları iyi çalışmayabiliyor. Bunun için sayısallaştırıyoruz verilerimizi.

DİKKAT!
Sayısallıştırılmış verileri işlerken dikkatli olmak lazım, makine öğrenmesi, kümeleme gibi algoritmalara bu verileri verdiğimizde, sayısal oldukları için farklı değerlendirilebilirler. 
ÖRNEK: Karşılaştırdığınız iki kişinin TC Kimlik numarasının biribirinden büyük veya küçük olması size bir bilgi vermez. Aynı şekilde taşıtların Sınıf Numarasının büyük veya küçük olmasına göre bir kıyaslama yapmamak gerekir. Algoritmalara tam hakim olmadan kullandığımızda bu bilgi sinsicene bilgi olarak sızabilir.
    
Makine öğrenmesinde kullanacaksak bu verileri "one hot" encoding kullanarak sayısallaştırmayı tercih edilmeliyiz.


Teker teker bakmak zor oluyor. Bir döngüye alalım, hepsini göstersin bize.

```python
plt.rcParams['figure.figsize'] = [5, 5]
for e in range(1,arabalar.MakeId.max()+1):
    ax3 = arabalar[arabalar.MakeId ==e].plot.scatter(x='FUELCONSUMPTION_COMB',y='CO2EMISSIONS', c='CYLINDERS', s=size,alpha=0.5,colormap='viridis')
    ax3.legend(str(e))
    
    plt.show()
```
