matplotlib'in  FuncAnimation ve PillowWriter modülleri sayesinde animasyonlar üretmek mümkün.

```python

    import numpy as np  
    import matplotlib.pyplot as plt  
    from matplotlib.animation import FuncAnimation, PillowWriter  
    %matplotlib notebook  
```

FuncAnimation aşağıdaki argümanları alıyor.

    fig -  Üzerine çizilecek nesne
    func - Her bir satırı çizmek için çağrılacak fonksiyon
    frames - fonksiyona verilecek verinin kaynağı
    init_func - Başlangıçta çalıştırılacak fonksiyon

%matplotlib notebook  - magic command  - notebook ortamında görsellerimizin interaktif olmasını sağlıyor.
 
Başlangıçta boş nesnelerimizi oluşturalım.

```python


    fig, ax = plt.subplots()  #grafigimiz ve eksenleri
    x, ysin, ycos = [], [], []  # gorselleyecegimiz veriler
    ln1, = plt.plot([], [], 'ro')  # yuvarlaklardan olusan cizgi
    ln2, = plt.plot([], [], 'm*')  # yildizlardan olusan cizgi

```


İlk değerleri atama, ve güncelleme fonksiyonlarını tanımlayalım.

```python

    def init():  
        ax.set_xlim(0, 2*np.pi)  # grafigin limitlerini X
        ax.set_ylim(-1, 1)      # grafigin limitleri Y
      
    def update(i):  
        x.append(i)               # yeni gelen veriyi listeye ekle
        ysin.append(np.sin(i))    # yeni gelen verileri listeye ekle
        ycos.append(np.cos(i))  
        ln1.set_data(x, ysin)      # listeleleri grafige veri olarak besle
        ln2.set_data(x, ycos)  
```

Animasyonu olusturalım:

```python

    ani = FuncAnimation(fig, update, np.linspace(0, 2*np.pi, 64), init_func=init)  
    # kaydetmek icin
    writer = PillowWriter(fps=25)  # yazdirma, kac kare / saniye
    ani.save("demo_sine.gif", writer=writer)  # dosyaya kaydedelim

    
    plt.show() # yukarida olusturdugumuz grafigi tetikler
```

0 - 6.28 arasında 64 tane değeri fonksiyonumuza veri olarak veriyoruz. (Bunlar sirayla update fonksiyonuna veriliyor)
